```python
import matplotlib.pyplot as plt

# Function to display ethnicity options and get user selection
def select_ethnicity():
    print("\nPlease select your ethnicity from the options below:")
    print("1. Asian or Asian British")
    print("2. Black, African, Caribbean or Black British")
    print("3. Middle Eastern")
    print("4. Mixed or multiple ethnicities with an Asian, Black or Middle Eastern background")
    print("5. White")
    print("6. Other ethnic group")
    print("7. Prefer not to say")
    
    # Get user's choice and validate input
    while True:
        try:
            choice = int(input("Enter the number corresponding to your ethnicity: "))
            if choice in range(1, 8):
                ethnicity = [
                    "Asian or Asian British",
                    "Black, African, Caribbean or Black British",
                    "Middle Eastern",
                    "Mixed or multiple ethnicities with an Asian, Black or Middle Eastern background",
                    "White",
                    "Other ethnic group",
                    "Prefer not to say"
                ][choice - 1]
                return ethnicity
            else:
                print("Please enter a valid number between 1 and 7.")
        except ValueError:
            print("Invalid input! Please enter a number between 1 and 7.")

# Function to calculate BMI, provide assessment, and plot BMI chart
def calculate_bmi():
    # User input for name
    name = input("Please enter your name: ")
    
    # Get ethnicity
    ethnicity = select_ethnicity()
    
    try:
        # User input for weight and height
        weight = float(input("\nEnter your weight in kilograms: "))
        height_cm = float(input("Enter your height in centimeters: "))
        
        # Convert height from centimeters to meters
        height_m = height_cm / 100
        
        # Check for valid inputs
        if weight <= 0 or height_m <= 0:
            print("Weight and height must be positive values.")
            return
        
        # BMI calculation
        bmi = weight / (height_m ** 2)
        
        # Display the BMI result with explanation, two decimal points, and ethnicity
        print(f"\n{name}, Body Mass Index (BMI) is calculated as your weight divided by the square of your height, or BMI = weight/height².")
        print(f"Your calculated BMI is: {bmi:.2f}")
        print(f"Ethnicity: {ethnicity}")
        
        # Determine the BMI category and give tailored advice
        if bmi < 15:
            category = "Underweight"
            advice = "You are underweight. A balanced diet rich in nutrients may help you achieve a healthier weight."
        elif bmi <= 24.9:
            category = "Normal weight"
            advice = "You are normal weight. Maintain your healthy lifestyle!"
        elif bmi < 29.9:
            category = "Overweight"
            advice = "You are overweight. Regular exercise and a balanced diet can help reach a healthier BMI."
        elif bmi < 34.9:
            category = "Obese"
            advice = "You are obese. It’s advisable to consult a healthcare provider for a personalized health plan."
        elif bmi < 39.9:
            category = "Severely obese"
            advice = "You are severely obese. Consider seeking guidance from a healthcare professional for weight management."
        else:
            category = "Morbidly obese"
            advice = "You are morbidly obese. A healthcare provider can provide support and options for weight management."
        
        # Display advice
        print(advice)
        
        # Plotting the BMI category with color scale
        categories = ["Underweight", "Normal weight", "Overweight", "Obese", "Severely obese", "Morbidly obese"]
        category_ranges = [15, 24.9, 29.9, 34.9, 39.9, 50]  # Upper limits for each category
        colors = ["lightblue", "green", "yellow", "orange", "orangered", "darkred"]

        plt.figure(figsize=(10, 6))
        bars = plt.bar(categories, category_ranges, color=colors, edgecolor='black')
        
        # Add a line to indicate the user's BMI
        plt.axhline(y=bmi, color='blue', linestyle='--', label=f"Your BMI: {bmi:.2f}")
        
        # Customize the plot
        plt.title(f"BMI Chart for {name} ({ethnicity})")
        plt.xlabel("BMI Categories")
        plt.ylabel("BMI Range")
        plt.legend()
        plt.show()
    
    except ValueError:
        print("Invalid input! Please enter numbers only for weight and height.")

# Run the BMI calculator
calculate_bmi()
```
